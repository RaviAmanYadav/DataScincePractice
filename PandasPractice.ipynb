# Feature Engginering

```python
loan3 = pd.read_csv("./Data/loan_data_set.csv")
```

```python
loan3
```

```python
loan3["total_income"]=loan3["ApplicantIncome"]+loan3["CoapplicantIncome"]
```

```python
loan3.head()
```

```python
loan3["LoanAmount"]=loan3["LoanAmount"].fillna(loan3["LoanAmount"].median())
```

```python
loan3["LTI"]=(loan3["LoanAmount"]/loan3["total_income"])*100
```

```python
loan3.head()
```

```python
loan3["LTI"].describe()
```

```python
def income(var):
    if var<5000:
        return "Low Income grp"
    elif var>=5000 and var<1000:
        return "mid income grp"
    else:
        return "High income grp"
```

```python
loan3["Binning"]=loan3["total_income"].apply(income)
```

```python
loan3.head(10)
```

```python
loan3["LoanAmount"].describe()
```

```python
# column for stability and unstability
def stable(var):
    if (var["Married"]=="Yes") & (var["Education"]=="Graduate") & (var["Self_Employed"]=="No"):
        return "stable"
    else:
        return "unstable"
```

```python
loan3["Stability"]=loan3.apply(stable,axis=1)
```
