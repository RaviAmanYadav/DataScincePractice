```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load data
data = sns.load_dataset('tips')
data.head()
```

```python
# Line Plot
sns.lineplot(x="total_bill", y="tip", data=data)
plt.title("Bill v/s Tip")
plt.show()
```

```python
# Bar plot
sns.barplot(x="day", y="total_bill", data=data)
plt.title("Average Bill per day")
plt.show()
```

```python
# Count plot
sns.countplot(x="day",data=data)
plt.title("Count Day")
plt.show()
```

```python
# box plot
sns.boxplot(x="day", y="total_bill", data=data)
plt.title('Bill generate per day')
plt.show()
```

```python
# heat map
tips = sns.load_dataset("tips")


#pivot table
pivot_table = tips.pivot_table(values="total_bill",index="day", columns="time", aggfunc="mean")

# Create heatmap
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu")

# Add title
plt.title("Average Total Bill by Day and Time")
plt.show()
```
