```python
import pandas as pd
import seaborn as sns
```

```python
# importing files
df = pd.read_csv("https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv")
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.info()
```

```python
pd.set_option('display.max_columns', None)
df.describe(include="all").round(2)
```

```python
# Find count of unique values in location column
df['location'].nunique()
```

```python
# find which continent has maximum frequency using values counts.
df['continent'].value_counts()
```

```python
# Find maximum & mean value in 'total_cases'
print("Total cases max value:",df['total_cases'].max())
print("Total cases mean value:", df['total_cases'].mean())
```

```python
#  Find 25%,50% & 75% quartile value in 'total_deaths'
df['total_deaths'].quantile([0.25, 0.5,0.75])
```

```python
# Find which continent has maximum 'human_development_index'.
df.describe()
```

```python
df.groupby("continent").agg({"human_development_index":"max"}).head(1)
```

```python
# Find which continent has minimum 'gdp_per_capita'.
df.groupby("continent").agg({"gdp_per_capita":"min"}).head(1)
```

```python
# Filter the dataframe with only this columns: ['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] and update the data frame.
df = df[["continent","location","date","total_cases","total_deaths","gdp_per_capita","human_development_index"]]
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
# finding the missing value
df.isnull().sum()
```

```python
# remove all the observation where continent is null
df.dropna(subset=["continent"], inplace=True)
df
```

```python
df = df.fillna(0)
```

```python
df.isnull().sum()
```

```python
# Convert date column in datetime format using pandas.to_datetime
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

```python
df["months"] = pd.DatetimeIndex(df["date"]).month
df.head()
```

```python
#  Find max value in all columns using groupby function on 'continent' column
df.groupby("continent").max().reset_index()
```

```python
df_groupby = df.groupby("continent").max().reset_index()
df_groupby
```

```python
# Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'
df_groupby["total_deaths_to_total_cases"]=df_groupby["total_deaths"]/df_groupby["total_cases"]
df_groupby["total_deaths_to_total_cases"]*100
df_groupby.head()
```

```python
sns.distplot(df['gdp_per_capita'])
```

```python
# Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
```

```python
sns.scatterplot(x="total_cases",y="gdp_per_capita",data=df)
```

```python
# Plot Pairplot on df_groupby dataset.
sns.pairplot(df_groupby)
```

```python
# Plot a bar plot of 'continent' column with 'total_cases' . Tip : using kind='bar' in seaborn catplot
sns.catplot(data=df_groupby,x="continent",y="total_cases", kind="bar")
```
