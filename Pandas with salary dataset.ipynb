```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\AC\Desktop\Notes\salary_data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(14, 6))  # adjust graph height and width
sns.barplot(x='YearsExperience', y='Salary', data=data)
plt.title("YearsExperience V/S Salary")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='YearsExperience', y='Salary', data=data)
plt.show()
```

```python
sns.scatterplot(x='YearsExperience', y='Salary', data=data)
plt.show()
```

```python
# point plot
plt.figure(figsize=(14, 6))
sns.pointplot(x='YearsExperience', y='Salary', data=data)
plt.xticks(rotation=45)
plt.show()
```
