```python
data.drop(['deck'], axis=1, inplace=True)
```

```python
data.head()
```

```python
data['age'].fillna(data['age'].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
data['embarked'].fillna(data['embarked'].mode()[0], inplace=True)
```

```python
data['embark_town'].fillna(data['embark_town'].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

## 3. Feature engineering

```python
data.head()
```

```python
data['family_size'] = data['sibsp'] + data['parch'] + 1
```

```python
data.head()
```

```python
data['age_group'] = pd.cut(
    data['age'],
    bins = [0, 12, 18, 35, 60, 100],
    labels = ['Child', 'Teeagers', 'Young Adult', 'Adult', 'Senior']
)
```

```python
data.head()
```

## 4. Filtering & Grouping

```python
data.groupby('sex')['survived'].mean()
data.groupby(['pclass', 'sex'])['survived'].mean().unstack()
```

## 5. Visualization Practice

```python
plt.figure(figsize=(8, 5))
plt.hist(data['age'], bins=20, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=data)
sns.boxplot(x='pclass', y='age', data=data)
sns.violinplot(x='sex', y='fare', data=data)
sns.barplot(x='embark_town', y='survived', data=data)

sns.pairplot(data, hue='sex', vars=['age', 'fare', 'pclass'])
```

## Pivot Table

```python
pivot = data.pivot_table(
    values='survived',
    index='pclass',
    columns='sex',
    aggfunc='mean'
)
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
```

```python
sns.barplot(x='embark_town', y='survived', data=data)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14,10))

sns.countplot(x='sex', hue='survived', data=data, ax=axes[0,0])
axes[0,0].set_title("Survival Count by Gender")

sns.boxplot(x='pclass', y='age', data=data, ax=axes[0,1])
axes[0,1].set_title("Age Distribution by Class")

sns.violinplot(x='sex', y='fare', data=data, ax=axes[1,0])
axes[1,0].set_title("Fare Distribution by Gender")

sns.barplot(x='embark_town', y='survived', data=data, ax=axes[1,1])
axes[1,1].set_title("Survival Rate by Embark Town")

plt.tight_layout()
plt.show()
```
