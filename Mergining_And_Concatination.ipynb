```python
import pandas as pd
```

```python
Dataset_1=pd.DataFrame({
    "roll_number":[1,2,3,4,5],
    "First_name":["Aman","Anchal","Pushpa","Rama","Mayank"],
    "Last_name":["X","Y",'Z','A','B'],
    "English":[90,54,73,98,87]
})
```

```python
Dataset_2=pd.DataFrame({
    "roll_number":[6,7,8,9,10],
    "First_name":["Ravi","Annu","Pushpa","Rama","Mayank"],
    "Last_name":['X2','Y2','Z2','A2','B2'],
    "Hindi":[45,74,83,88,97]
})
```
