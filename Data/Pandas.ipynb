# Feature Engineering:

```python
df1
```

```python
#total income
df1["total_income"]=df1["ApplicantIncome"]+df1["CoapplicantIncome"]
```

```python
df1
```

```python
#loan to income ratio:
df1["LTI"]=(df1["LoanAmount"]/df1["total_income"])*100
```

```python
df1
```

```python
df1["LTI"].describe()
```

```python
#total income as <5000: low income
ti>5000 and ti<100000: avg
    >10000 high
```

```python
def income(var):
    if var<5000:
        return "Low income group"
    elif var>=5000 and var<10000:
        return "mid income group"
    else:
        return "High Income group"
```

```python
df1["Binning"]=df1["total_income"].apply(income)
```

```python
df1
```

```python
#column for stability and instability

married,graduated and not self employed----->stable
else unstable
```

```python
def stable(var):
    if (var["Married"]=="Yes")& (var["Education"]=="Graduate") &(var["Self_Employed"]=="No"):
        return "Stable"
    else:
        return "Unstable"
```

```python
df1["Stability"]=df1.apply(stable,axis=1)
```

```python
df1
```
